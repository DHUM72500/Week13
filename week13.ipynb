```python
# import wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Define text
twocities = "It was the best of times, it was the worst of times, it was the age of innocence it was the age of expreience, it was really really bad. "

# Define Word Cloud object
cloud_twocities = WordCloud().generate(twocities)
```

```python
# The Word Cloud function has many arguments. To find the arguments, type ?WordCloud a window will pop up
# below that you can use to find the argument and attribute that you'd like to modify. 

?WordCloud
```

```python
# What we've done so far is create a word cloud object. You can see what it is by calling it below.

cloud_twocities
```

```python
#How does the Tale of Two Cities WordCloud look? 
# To display--> must use Matplotlib functionality. There are 3 plt steps

plt.imshow(cloud_twocities, interpolation='bilinear')

# imshow creates the picture, but we need to call "show" to display it. the first argument names the cloud object and the second is an attribute
# turn the axis off because word clouds are not plotted on an axis. 
plt.axis('off')

# Then show it. 
plt.show()
```

```python
# Building a wordcloud

my_cloud = WordCloud(background_color= 'white').generate(twocities)

plt.imshow(my_cloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
# This example includes removing a standard set of stopwords from the dataset.

from wordcloud import WordCloud, STOPWORDS

# define stopwords list
my_stopwords = set(STOPWORDS)


my_cloud = WordCloud(background_color = 'white', stopwords = my_stopwords).generate(twocities)
plt.imshow(my_cloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()

# You can also remove stopwords using the CountVectorizer function. 

```
